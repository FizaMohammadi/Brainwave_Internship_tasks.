```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Step 1: Create an enriched dataset

# Step 1: Create an enriched dataset

data = {
    "Date": pd.date_range(start="2024-10-01", periods=20),
    "Invoice ID": [f"INV{1000 + i}" for i in range(20)],
    "Customer ID": [f"CUST{i%5}" for i in range(20)],  # Repeating customers
    "Product": ["Mobiles", "Laptop", "Refrigerator", "Car"] * 5,
    "Category": ["Electronics", "Gadget", "Home", "Automobile"] * 5,
    "Price": [25000, 70000, 50000, 800000] * 5,
    "Quantity Sold": [10, 20, 15, 8, 12, 18, 5, 22, 14, 11, 9, 7, 30, 25, 6, 13, 19, 17, 11, 2],
    "Region": ["North", "South", "East", "West"] * 5,
    "Salesperson": ["Alice", "Bob", "Charlie", "David", "Emily"] * 4,
    "Discount": np.random.randint(10, 25, size=20),
    "Shipping Cost": np.random.uniform(5, 20, size=20),
    "Profit Margin": np.random.uniform(0.2, 0.4, size=20),
    "Warranty Period (Years)": [2, 3, 1, 5] * 5,
    "Brand": ["Samsung", "Dell", "LG", "Toyota"] * 5,
    "Power Consumption (Watts)": [10, 50, 200, 0] * 5,  # Cars usually donâ€™t have this field
    "Battery Life (Hours)": [24, 12, None, None] * 5,  # Not all products have batteries
    "Safety Rating (Out of 5)": [4.5, 4.0, 3.5, 5.0] * 5,
    "Technology/Feature": ["5G Enabled", "AI Powered", "Smart Cooling", "Hybrid"] * 5,
    "Service Centers (Region)": [20, 25, 15, 30] * 5,
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
df["Total Sales"]=df["Price"]*df["Quantity Sold"]
```

```python
df["Battery Life (Hours)"]=df["Battery Life (Hours)"].fillna(df["Battery Life (Hours)"].mode()[0])
```

```python
# Calculate total sales by region
total_sales_by_region = df.groupby('Region')['Total Sales'].sum()
```

```python
# Visualize the results
plt.figure(figsize=(10, 6))
total_sales_by_region.plot(kind='bar')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

```python
# 2. Top-Selling Products by Category:
# Group by category and sum the quantities sold
top_selling_products = df.groupby(['Category', 'Product'])['Quantity Sold'].sum().sort_values(ascending=False)
```

```python
# Print the top 5
print(top_selling_products.head(5))
```

```python
#3. Salesperson Performance:
# Group by salesperson and calculate total sales
salesperson_performance = df.groupby('Salesperson')['Total Sales'].sum()

# Visualize the performance
plt.figure(figsize=(10, 6))
salesperson_performance.plot(kind='bar')
plt.title('Salesperson Performance')
plt.xlabel('Salesperson')
plt.ylabel('Total Sales')
plt.show()
```

```python
#4. Product-Wise Profit Margin:
# Group by product and calculate average profit margin
product_profit_margin = df.groupby('Product')['Profit Margin'].mean()

# Visualize the profit margins
plt.figure(figsize=(10, 6))
product_profit_margin.plot(kind='bar')
plt.title('Product-Wise Profit Margin')
plt.xlabel('Product')
plt.ylabel('Average Profit Margin')
plt.show()
```

```python
#5. Time Series Analysis:
# The 'Date' column is already the index, no need to set it again
# df.set_index('Date', inplace=True)  # Remove or comment out this line

# Plot the total sales over time
plt.figure(figsize=(10, 6))
df['Total Sales'].plot()
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()
```

```python
# 1. Total Sales by Product
df['Total Sales'] = df['Price'] * df['Quantity Sold']
total_sales_by_product = df.groupby('Product')['Total Sales'].sum()

# 2. Sales by Region
sales_by_region = df.groupby('Region')['Total Sales'].sum()

# 3. Top Selling Products (by quantity sold)
top_selling_products = df.groupby('Product')['Quantity Sold'].sum().sort_values(ascending=False)

# 4. Profit Margin and Discount Analysis
df['Discounted Price'] = df['Price'] - (df['Price'] * df['Discount'] / 100)
profit_margin_vs_discount = df.groupby('Discount')['Profit Margin'].mean()

# 5. Sales Performance by Salesperson
sales_by_salesperson = df.groupby('Salesperson')['Total Sales'].sum()

# 6. Warranty and Product Categories
warranty_by_category = df.groupby('Category')['Warranty Period (Years)'].mean()

# 7. Shipping Costs Analysis
shipping_cost_by_product = df.groupby('Product')['Shipping Cost'].mean()

# Display insights
print("Total Sales by Product:\n", total_sales_by_product)
print("\nSales by Region:\n", sales_by_region)
print("\nTop Selling Products (by Quantity Sold):\n", top_selling_products)
print("\nProfit Margin vs Discount:\n", profit_margin_vs_discount)
print("\nSales by Salesperson:\n", sales_by_salesperson)
print("\nWarranty by Product Category:\n", warranty_by_category)
print("\nShipping Cost by Product:\n", shipping_cost_by_product)

```
